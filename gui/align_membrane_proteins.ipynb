## Align membrane proteins

This notebook assits to align a macromlecule with respect to a reference membrane. First, a reference membrane an Z-axis is constructed. Second, the user aligns in 3D the protein with respect the reference membrane and Z-axis.m

```python
from core.vtk_utilities import *
from core.utilities import *
from core.widgets_utilities import *
from core.tk_utilities import *
import io
```

## Create reference volume with membrane and Z-axis

Generate cubic density maps to work as reference to align membrane proteins. This notebook starts from MRC files like those generated by *atomic_to_density.ipynb*:

- List of files you want to create an axis reference, more than one file can be selected by using "Ctrl" key:
- Membrane size: is used to define the size of the membrane region in Angstroms.
- Zaxis rad: define the radius of the cylindrical region in angstroms around the Z axis.
- Pixels size: as explained above the scale of the tomogram resolution.
- Select the path you want to save the file or files.

Afther the values are selected click on **Save** button to exec the script.

```python
PATH = "../data/templates/mb_mrcs_10A/"
files_path = []

def select(btn):
    global files_path
    select_files(files_path, ".mrc")
    replace_options(files_path, dropdown)

def create_axis(btn):
    global files_path
    with output_widget:
        path = check_dir(file_destination_widget.value , PATH)
        create_axis_mrc(mb_size_widget.value, zaxis_rad_widget.value, vsize_widget.value, files_path, path, scale_factor_widget.value)
        files_path = [] 
        replace_options(files_path, dropdown)
    
(mb_size_widget, zaxis_rad_widget, vsize_widget, scale_factor_widget, select_files_button, dropdown, save_button, file_destination_widget) = widgets_create_axis()
save_button.on_click(create_axis)
select_files_button.on_click(select)
output_widget = widgets.Output()
display(output_widget)
```

## Align membrane proteins

The membrane protein has to be manually position on the reference membran and Z-axis created above.

Some parameters must be adjusted before visualization:

- Select membrane file you want to align, a window will appear to select the file.
- Select the axis file you want to use to align the protein.
- Select the path and the ouput name you want to put in the new file. If you don't add any, the default path and name will be used.
- voxel size of the density map important to be the same that the input map.
- Press **Exec** to align the files.

A VTK window will be opened with the protein and the membrana reference with Z-axis. For protein movement use the VTK interactor switch which have four controls:

- Key j: the joystick actor will active.
- Key t: the joystick camera will active.
- key a: the trackball actor will active.
- key c: the trackball camera will active.
- key s: press it to save the membrane position over the axis. This will create a new density map with the aligned protein.

Recommendation: use 't' for orientation and moving the object.

Repeat this proccess for every membrane protein you want to use in the application.

```python
DEFAULT_PATH2 = "../data/templates/mb_mrcs_10A/"
file_path = ""
file_path2 = ""

def select_membrane(btn):
    global file_path
    file_path = select_file(".mrc")
    dropdown_1.options = [file_path]
    dropdown_1.value = file_path
    
def select_axis(btn):
    global file_path2
    file_path2 = select_file(".mrc")
    dropdown_2.options = [file_path2]
    dropdown_2.value = file_path2

def show_window(btn):
    with output_widget_2:
        path = check_dir(file_destination_widget.value, DEFAULT_PATH2)
        protein_to_axis(file_path, file_path2, 800, 600, vsize2_widget.value, path)

upload_protein_button, upload_axis_button, exec_button, dropdown_1, dropdown_2, vsize2_widget,file_destination_widget = widgets_align()
exec_button.on_click(show_window)
upload_protein_button.on_click(select_membrane)
upload_axis_button.on_click(select_axis)
output_widget_2 = widgets.Output()
display(output_widget_2)
```
